# Perceptron Learning Algorithm

```python
import numpy as np
import pandas as pd
import random
```

```python
#Read the dataset 'classification',remove the last column:
dataset = pd.read_table('classification.txt',sep=',',header=None,names=['x1','x2','x3','label','na'])
dataset = dataset.drop(['na'],axis=1)

#Add the first column-X0:
x0 = []
for i in range(len(dataset)):
    x0.append(1)

first_col = pd.DataFrame(columns=['x0'],data=x0)

dataset = first_col.merge(dataset,how='inner',left_index=True, right_index=True)
```

```python
#Check our dataset:
dataset
```

```python
#Divide our dataset into X(coordinates of points) and Y(label):
X = dataset.drop(['label'],axis=1)
Y = dataset.loc[:,'label']  
```

```python
#Randomize the initial weights:
weights = np.random.random(4)
print('The initial weights are',weights)
```

```python
#Define PLA function to update weights:
def perceptron_learning_algo (data,x,y,w,n_iteration,alpha):
    n = 0
    while n < n_iteration:
        n+=1 
        for i in range(len(data)):
            thershold = np.dot(w,(np.array(x.iloc[i]))) #Calculate the thershold function
            if thershold < 0 and y.iloc[i] == 1: #Pick the violated data points and update weights
                w = w + alpha *np.array(x.iloc[i])
                break
            elif thershold >= 0 and y.iloc[i] == -1:
                w = w - alpha *np.array(x.iloc[i])
                break
            else:
                continue 
                              
    print('The weights after PLA iterates',n_iteration,'times are',w)  
    return w 
```

```python
#Define the condition for iteration:
#n_iteration = 7000
#alpha = 0.01

#update the weights 7000 tims with alpha 0.01:
weights_updated=perceptron_learning_algo(dataset,X,Y,weights,7000,0.01)
```

```python
#Add the predicted results to our dataset:
predict = []

for i in range(len(dataset)):
    if np.dot(weights_updated,(np.array(X.iloc[i]))) >= 0:
        predict.append(1)
    elif np.dot(weights_updated,(np.array(X.iloc[i]))) < 0:
        predict.append(-1)

predict_col = pd.DataFrame(columns=['Predicted label'],data=predict)
new_dataset = dataset.merge(predict_col,how='inner',left_index=True, right_index=True)
```

```python
#Calculate the model accuracy rate:
accurate = 0
for i in range(len(new_dataset)):
    if new_dataset.iloc[i][4] == new_dataset.iloc[i][5]:
        accurate+=1
    else:
        pass

accuracy_rate = accurate/(len(new_dataset))
print ('The accuracy of PLA after model iterates 7000 times is',accuracy_rate)
```
